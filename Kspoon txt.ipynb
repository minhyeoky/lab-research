# Text 전처리2

```python
import re
```

```python
def _prep_txt(txt):
    """
    (철자전사)/(발음전사)로 표기된 경우 발음전사 사용

    b/ : 숨소리 -> 제거
    l/ : 웃음소리 -> 데이터 제거
    o/ : 다음사람 -> 데이터 제거
    n/ : 잡음 -> 제거
    u/ : 전혀 알아들을 수 없음 -> 데이터 제거

    + : 반복 발성 -> 모두 사용, '+'만 제거
    * : 발음이 부정확 -> 제거
    +, * 가 같이 있는 경우, 기호 제거 후 사용
    """
    pattern = '[(]{1}[a-z가-힣0-9 ]+[)]{1}[\/]{1}[(]{1}[a-z가-힣0-9 ]+[)]{1}'
    # (철자전사)/(발음전사)
    delete_list = ['l/', 'o/']
    remove_list = ['b/', 'n/', '+', '*', '*+', '+*', '/', '\n', '.', '?', '!', ',']
    for each in delete_list:
        if each in txt:
            return False
    regex = re.compile(pattern)
    for each in regex.findall(txt):
        rep = each.split('/')
        rep = rep[-1][1:-1]
        txt = txt.replace(each, rep)
    for each in remove_list:
        txt = txt.replace(each, '')
    return ' '.join(txt.strip().split()) # white space 정리
```

```python
idx = 0
for each in Path('../data/KsponSpeech_01').iterdir():
    for file in each.glob('*.txt'):
        try:
            with open(str(file), mode='r', encoding='euc_kr') as f:
                txt = f.readline().strip('\n')
        except UnicodeDecodeError:
            with open(str(file), mode='r', encoding='cp949') as f:
                txt = f.readline().strip('\n')
        finally:
            with open(f'{file}.prep', mode='w', encoding='utf8') as f:
                txt = _prep_txt(txt)
                if not txt:
                    pass
                else:
                    if idx % 500 == 0:
                        print(txt)
                    idx += 1
                    f.write(_prep_txt(txt))
```
