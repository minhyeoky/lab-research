# Text 전처리2

```python
import re
```

```python
def _prep_txt(txt):
    """
    (철자전사)/(발음전사)로 표기된 경우 발음전사 사용

    b/ : 숨소리 -> 제거
    l/ : 웃음소리 -> 데이터 제거
    o/ : 다음사람 -> 데이터 제거
    n/ : 잡음 -> 제거
    u/ : 전혀 알아들을 수 없음 -> 데이터 제거

    + : 반복 발성 -> 모두 사용, '+'만 제거
    * : 발음이 부정확 -> 제거
    +, * 가 같이 있는 경우, 기호 제거 후 사용
    """
    pattern = '[(]{1}[a-z가-힣0-9 ]+[)]{1}[\/]{1}[(]{1}[a-z가-힣0-9 ]+[)]{1}'
    # (철자전사)/(발음전사)
    delete_list = ['l/', 'o/']
    remove_list = ['b/', 'n/', '+', '*', '*+', '+*', '/', '\n', '.', '?', '!', ',']
    for each in delete_list:
        if each in txt:
            return False
    regex = re.compile(pattern)
    for each in regex.findall(txt):
        rep = each.split('/')
        rep = rep[-1][1:-1]
        txt = txt.replace(each, rep)
    for each in remove_list:
        txt = txt.replace(each, '')
    return ' '.join(txt.strip().split()) # white space 정리
```

```python
txt = '내가 그양 적응 못 할 때 일부러 좀 막 데꼬 나가고 많이 뭐 먹을 거 사주고 이래가지고'

txt = txt.strip().split()
```

```python
# https://github.com/CyberZHG/keras-bert/blob/master/keras_bert/tokenizer.py
# wordpiece tokenize
tokens = []
for each in txt:
    if each in dl.vocab:
        tokens.append(each)
        continue
    
    start = 0
    end = 0
    while start < len(each):
        end = len(each)
        while end > start:
            sub = each[start:end]
            if start > 0:
                sub = '##' + sub
            if sub in dl.vocab:
                break
            end -= 1
        if start == end:
            # 못찾음
            end += 1
        tokens.append(sub)
        start = end
        
```
